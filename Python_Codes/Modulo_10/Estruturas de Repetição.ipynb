#### 2. Faça um programa que leia um nome de usuário e a sua senha e não aceite a senha igual ao nome do usuário, mostrando uma mensagem de erro e voltando a pedir as informações.

````python
usuario = input('Digite o usuário: ')
senha = input('Digite a senha: ')

while usuario == senha:
    print('A senha não pode ser igual ao usuário!')
    usuario = input('Digite o usuário: ')
    senha = input('Digite a senha')

print('Usuário: {} - Senha: {}'.format(usuario.capitalize(), senha))
````

#### 4. Supondo que a população de um país A seja da ordem de 80000 habitantes com uma taxa anual de crescimento de 3% e que a população de B seja 200000 habitantes com uma taxa de crescimento de 1.5%: Faça um programa que calcule e escreva o número de anos necessários para que a população do país A ultrapasse ou iguale a população do país B, mantidas as taxas de crescimento.

````python
pop_a = 80000
pop_b = 200000
tempo = 0

while pop_a <= pop_b:
    pop_a *= 1.03
    pop_b *= 1.015
    tempo += 1

print('Serão necessários {} anos para a população A alcançar a população B'.format(tempo))
````

#### 6. Faça um programa que peça para o usuário inserir o faturamento dos últimos 5 meses (individualmente) e informe o maior faturamento

````python
print('Faturamento dos últimos 5 meses')
print('===============================')
print('')

for i in range(5): 
    faturamento = float(input('Informe o faturamento do mês {}:'.format(i + 1)))
    print('Faturamento do mes {}: ${:,.2f}'.format((i + 1), faturamento))
    if i == 0:
        maior_faturamento = faturamento
    else:
        if faturamento > maior_faturamento:
            maior_faturamento = faturamento

print('')
print('O maior faturamento foi de ${:,.2f}'.format(maior_faturamento))
````

#### 7. Faça um programa que peça para o usuário inserir o faturamento dos últimos 5 meses (individualmente) e informe o faturamento total (soma) e o faturamento médio por mês (média).

````python
print('Faturamento dos últimos 5 meses')
print('===============================')

soma_faturamento = 0

for i in range(5): 
    faturamento = float(input('Informe o faturamento do mês {}:'.format(i + 1)))
    soma_faturamento += faturamento
    print('Faturamento do mes {}: ${:,.2f}'.format((i + 1), faturamento))
    
print('===============================')
print('O faturamento total foi de ${:,.2f}'.format(soma_faturamento))
print('A média de faturamento foi de ${:,.2f} por mês'.format(soma_faturamento / (i + 1)))
````

#### 8. Faça um programa que consiga categorizar a idade das equipes de uma empresa. Faça um programa que peça para n pessoas a sua idade. Ao final o programa, deverá verificar se a média de idade da equipe varia entre 0 e 25 (jovem) ,26 e 60 (sênior) e maior que 60 (idosa) e, então, dizer se a equipe é jovem, sênior ou idosa, conforme a média calculada.

````python
qtde_pessoas = int(input('Digite o número de pessoas: '))
#idade = 0
soma_idade = 0

print('Quantidade de pessoas: {} pessoas'.format(qtde_pessoas))

while qtde_pessoas <= 0:
    print('A quantidade de pessoas deve ser maior que 0')
    qtde_pessoas = int(input('Digite o número de pessoas: '))

for i in range(qtde_pessoas):
    idade = int(input('Digite a idade da {}ª pessoa: '.format(i + 1)))
    print('Idade da {}ª pessoa: {} anos'.format((i + 1), idade))
    soma_idade += idade

media_idade = soma_idade / qtde_pessoas

if 0 <= media_idade <= 25:
    print('A equipe é jovem, com uma média de idade de {:.1f} anos'.format(media_idade))
elif 26 <= media_idade <= 60:
    print('A equipe é sênior, com uma média de idade de {:.1f} anos'.format(media_idade))
else:
    print('A equipe é jovem, com uma média de idade de {:.1f} anos'.format(media_idade))
````

#### 9. Numa eleição existem três candidatos. Faça um programa que peça o número total de eleitores. Peça para cada eleitor votar e, ao final, mostre o número de votos de cada candidato.

````python
cand_1 = 0
cand_2 = 0
cand_3 = 0

qtde_eleitores = int(input('Digite a quanditade de eleitores: '))

while qtde_eleitores <= 0:
    print('A quantidade de eleitores deve ser maior que 0.')
    qtde_eleitores = int(input('Digite a quanditade de eleitores: '))

for i in range(qtde_eleitores):
    voto = int(input('Digite o número do candidato (1, 2 ou 3): '))
    while voto <= 0 or voto > 3:
        print('Candidato inválido. Digite o número do candidato (1, 2 ou 3): ')
        voto = int(input('Digite o número do candidato (1, 2 ou 3): '))

    if voto == 1:
        cand_1 += 1
    elif voto == 2:
        cand_2 += 1
    else:
        cand_3 += 1

print('Quantidade de eleitores: {} eleitores.' .format(qtde_eleitores))
print('Candidato 1: {} votos.'.format(cand_1))
print('Candidato 2: {} votos.'.format(cand_2))
print('Candidato 3: {} votos.'.format(cand_3))

````

#### 10. Faça um programa que calcule o valor total investido por um colecionador em sua coleção de CDs e o valor médio gasto em cada um deles. O usuário deverá informar a quantidade de CDs e o valor para em cada um.

````python
qtde_cd = int(input('Digite a quantidade de CDs: '))
valor_total = 0

while qtde_cd <= 0:
    print('A quanditade deve ser maior que 0.')
    qtde_cd = int(input('Digite a quantidade de CDs: '))

for i in range(qtde_cd):
    valor = float(input('Digite o valor do {}º CD: '.format(i + 1)))
    valor_total += valor

print('Quantidade de CDs no acervo: {} unidades.'.format(qtde_cd))
print('Valor total investido na coleção: ${:,.2f}'.format(valor_total))
print('Valor médio: ${:,.2f}'.format(valor_total / qtde_cd))
````

#### 12. Um funcionário de uma empresa recebe aumento salarial anualmente: Sabe-se que:
<pre>
Esse funcionário foi contratado em 1995, com salário inicial de R$ 1.000,00;
Em 1996 recebeu aumento de 1,5% sobre seu salário inicial;
A partir de 1997 (inclusive), os aumentos salariais sempre correspondem ao dobro do percentual do ano anterior.
Faça um programa que determine o salário desse funcionário em 2003. 
</pre>

````python
salario = 1000.0 * 1.015
porcent_aumento = 0.015

for i in range(1997, 2004):
    porcent_aumento *= 2
    salario *= (1 + porcent_aumento)

print('O salário em 2003 é de ${:,.2f}'.format(salario))
````

#### 13. O cardápio de uma lanchonete é o seguinte:
<pre>
Especificação   Código  Preço
Cachorro Quente 100     R$ 1,20
Bauru Simples   101     R$ 1,30
Bauru com ovo   102     R$ 1,50
Hambúrguer      103     R$ 1,20
Cheeseburguer   104     R$ 1,30
Refrigerante    105     R$ 1,00
Faça um programa que leia o código dos itens pedidos e as quantidades desejadas.
Calcule e mostre o valor a ser pago por item (preço * quantidade) e o total geral do pedido.
Considere que o cliente deve informar quando o pedido deve ser encerrado.
</pre>

````python
total = 0

codigo = input('Informe o código do produto. Para encerrar, digite Enter com o campo em branco: ')

while codigo.casefold() != '':
    qtde = int(input('Informe a quantidade: '))
    if codigo == '100':
        total += 1.2 * qtde
    elif codigo == '101':
        total += 1.3 * qtde
    elif codigo == '102':
        total += 1.5 * qtde
    elif codigo == '103':
        total += 1.2 * qtde
    elif codigo == '104':
        total += 1.3 * qtde
    elif codigo == '105':
        total += 1 * qtde
        
    codigo = input('Informe o código do produto. Para encerrar, digite Enter com o campo em branco: ')

print('O total a ser pago é: ${:,.2f}'.format(total))
````

#### 14. Uma grande emissora de televisão quer fazer uma enquete entre os seus telespectadores para saber qual o melhor jogador após cada jogo. Para isto, faz-se necessário o desenvolvimento de um programa, que será utilizado pelas telefonistas, para a computação dos votos. Sua equipe foi contratada para desenvolver este programa, utilizando a linguagem de programação Python. Para computar cada voto, a telefonista digitará um número, entre 1 e 23, correspondente ao número da camisa do jogador. Um número de jogador igual zero, indica que a votação foi encerrada. Se um número inválido for digitado, o programa deve ignorá-lo, mostrando uma breve mensagem de aviso, e voltando a pedir outro número. Após o final da votação, o programa deverá exibir:
<pre>a. O total de votos computados;
b. Os números e respectivos votos de todos os jogadores que receberam votos;
c. O percentual de votos de cada um destes jogadores;
d. O número do jogador escolhido como o melhor jogador da partida, juntamente com o número de votos e o percentual de votos dados a ele.
        Observe que os votos inválidos e o zero final não devem ser computados como votos. O resultado aparece ordenado pelo número do jogador. O programa deve fazer uso de arrays. O programa deverá executar o cálculo do percentual de cada jogador através de uma função. Esta função receberá dois parâmetros: o número de votos de um jogador e o total de votos. A função calculará o percentual e retornará o valor calculado. Abaixo segue uma tela de exemplo. O disposição das informações deve ser o mais próxima possível ao exemplo. Os dados são fictícios e podem mudar a cada execução do programa. Ao final, o programa deve ainda gravar os dados referentes ao resultado da votação em um arquivo texto no disco, obedecendo a mesma disposição apresentada na tela.</pre>
```
Enquete: Quem foi o melhor jogador?

Número do jogador (0=fim): 9
Número do jogador (0=fim): 10
Número do jogador (0=fim): 9
Número do jogador (0=fim): 10
Número do jogador (0=fim): 11

Número do jogador (0=fim): 50
Informe um valor entre 1 e 23 ou 0 para sair!
Número do jogador (0=fim): 9

Número do jogador (0=fim): 0

Resultado da votação:

Foram computados 8 votos.

Jogador Votos           %
9               4               50,0%
10              3               37,5%
11              1               12,5%
``` 

````python
votos = []

print('Enquete: Quem foi o melhor jogador?')

for i in range(23):
    votos.append(0)

while True:
    numero = int(input('Informe o número do jogador (digite 0 para finalizar): '))
    if numero == 0:
        break
    elif not 1 <= numero <= 23:
        print('Informe um valor entre 1 e 23 ou 0 para sair!')
        continue
    votos[numero - 1] += 1

total = sum(votos)
print('Foram computados {} votos'.format(total))
print('Jogador\tVotos\t%')
for i, voto in enumerate(votos):
    if voto > 0:
        print(f'{i+1}\t{voto}\t{voto/total:.1%}')

melhor = max(votos)
i = votos.index(melhor)
porcentagem = melhor/total
print('O melhor jogador foi o de número {}, com {} votos, correspondendo a {:.2%} do total de votos.'.format(i + 1, melhor, porcentagem))
````
